```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

print('Libraries loaded successfully.')
```

## 1. 產生合成資料 / Generate Synthetic Data

模擬工業場景：以設備運轉時數 (hours) 預測振動幅度 (vibration)。

```python
np.random.seed(42)

# 設備運轉時數 (0 ~ 1000 小時)
hours = np.linspace(0, 1000, 100)

# 振動幅度 (mm/s) — 真實關係: vibration = 0.05 * hours + 2 + noise
noise = np.random.normal(0, 1.5, size=hours.shape)
vibration = 0.05 * hours + 2 + noise

print(f'X shape: {hours.shape}, Y shape: {vibration.shape}')
print(f'Vibration range: {vibration.min():.2f} ~ {vibration.max():.2f} mm/s')
```

## 2. 視覺化資料 / Visualise Data

```python
plt.figure(figsize=(10, 5))
plt.scatter(hours, vibration, alpha=0.6, label='Sensor readings')
plt.xlabel('Operating Hours')
plt.ylabel('Vibration (mm/s)')
plt.title('Equipment Vibration vs Operating Hours')
plt.legend()
plt.grid(True)
plt.show()
```

## 3. 手動實作梯度下降 / Manual Gradient Descent

```python
def compute_cost(x, y, w, b):
    """Mean Squared Error cost function."""
    m = len(y)
    predictions = w * x + b
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    return cost


def gradient_descent(x, y, w_init, b_init, alpha, iterations):
    """Perform gradient descent to learn w and b."""
    m = len(y)
    w, b = w_init, b_init
    cost_history = []

    for _ in range(iterations):
        predictions = w * x + b
        dj_dw = (1 / m) * np.sum((predictions - y) * x)
        dj_db = (1 / m) * np.sum(predictions - y)
        w -= alpha * dj_dw
        b -= alpha * dj_db
        cost_history.append(compute_cost(x, y, w, b))

    return w, b, cost_history


w_final, b_final, costs = gradient_descent(
    hours, vibration, w_init=0.0, b_init=0.0, alpha=1e-6, iterations=1000
)
print(f'Learned w = {w_final:.4f}, b = {b_final:.4f}')
print(f'Final cost: {costs[-1]:.4f}')
```

## 4. 使用 scikit-learn 驗證 / Validate with scikit-learn

```python
X = hours.reshape(-1, 1)
model = LinearRegression()
model.fit(X, vibration)

print(f'sklearn — w = {model.coef_[0]:.4f}, b = {model.intercept_:.4f}')
y_pred = model.predict(X)
mse = mean_squared_error(vibration, y_pred)
print(f'MSE = {mse:.4f}')

plt.figure(figsize=(10, 5))
plt.scatter(hours, vibration, alpha=0.5, label='Data')
plt.plot(hours, y_pred, color='red', linewidth=2, label='Linear Regression')
plt.xlabel('Operating Hours')
plt.ylabel('Vibration (mm/s)')
plt.title('Linear Regression Fit')
plt.legend()
plt.grid(True)
plt.show()
```
